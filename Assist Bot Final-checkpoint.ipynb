```python
import speech_recognition as sr
import pyttsx3
import webbrowser
import tkinter as tk
import wikipedia
engine = pyttsx3.init()
voice_id = "HKEY_LOCAL_MACHINE\SOFTWARE\Microsoft\Speech\Voices\Tokens\TTS_MS_EN-US_ZIRA_11.0"
#female voice id
engine.setProperty('voice', voice_id) 

#some chats

greet=["hello","hai","hey"]
Owner=["Build","made","birth"]
wish={"morning":"Good morning","night":"Good night","care":"You too.","bye":"Please don't go"}
Comedy={"marry":"Why not!","love":"I love more","hate":"lmao, look at your face ","children":"Oh please."}

global answer
global order

def check():
        
    r=sr.Recognizer()
    with sr.Microphone() as source:
        audio=r.listen(source)  
        order=r.recognize_google(audio,language="en-in") #recognizing the audio
    e2.insert(tk.END,order) 
    global x
    x=order.split(' ') 
    
    for word in x:
        if word in greet:    
            answer="Hey, a pleasent day!"
            engine.say(answer)
            engine.runAndWait()
            e1.insert(tk.END, answer) 
            
        elif word in Owner:    
            answer="I am owned by Latish"
            engine.say(answer)
            engine.runAndWait()
            e1.insert(tk.END, answer) 
       
            
        elif word in Comedy:
            answer=Comedy[word]
            engine.say(answer)
            engine.runAndWait()
            e1.insert(tk.END, answer)    
       
            
        elif "search" in x:
            webbrowser.open_new(f'https:www.google.com/search?q={x[1]}')
            
        elif "what" in x or "who" in x:
            
            try:
                order=order.replace("who","")
        
            except:
                order=order.replace("what","")
            answer=wikipedia.summary(order,sentences=1)
            engine.say(answer)
            engine.runAndWait()
            e1.insert(tk.END, answer)
            break

            
        elif word in wish:
            answer=wish[word]
            engine.say(answer)
            engine.runAndWait()
            e1.insert(tk.END, answer)
            
        elif "quit" in x:
            root.destroy()
        else:
            answer="Sorry! Please come again"
            engine.say(answer)
            engine.runAndWait()
            e1.insert(tk.END, answer) 
            
            
            
#creating GUI
root=tk.Tk()
root.geometry("500x600")
root.title("Sai Latish")
root.configure(bg="green")
e1=tk.Text(root)
l1=tk.Label(root,text="Answer:")
l1.place(x=220,y=155)
l3=tk.Label(root,text="Welcome! Voice Bot By Latish")
l3.place(x=180,y=15)
e1.place(x=50,y=190,height=80,width=400)
e2=tk.Text(root)

l=tk.Label(root,text="Input:")
l.place(x=225,y=310)
e2.place(x=50,y=350,height=100,width=400)
b1=tk.Button(root,text="Speak",bg="red",fg="white",command=check)
b1.place(x=170,y=500,height=50,width=100)

root.mainloop()
```
